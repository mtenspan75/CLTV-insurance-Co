```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# defining X and Y
X = train_cee.drop('cltv' , axis = 1)
y = train_cee['cltv']

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Scale the data using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def handle_small_classes(X, y, threshold=5):
    class_counts = y.value_counts()
    small_classes = class_counts[class_counts < threshold].index
    mask = y.isin(small_classes)
    X_filtered = X[~mask]
    y_filtered = y[~mask]
    return X_filtered, y_filtered

X = train_cee.drop('cltv', axis=1)
y = train_cee['cltv']

# Filter out classes with less than 5 samples
X_filtered, y_filtered = handle_small_classes(X, y)

# Instantiate SMOTE with adjusted n_neighbors
n_neighbors = 3  # Choose a value suitable for your data
smote = SMOTE(random_state=42, k_neighbors=n_neighbors)

# Resample the data
X_resampled, y_resampled = smote.fit_resample(X_filtered, y_filtered)

# Scale the data using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_resampled)
X_test_scaled = scaler.transform(X_test)

# Perform train-test split on the resampled and scaled data
X_train, X_test, y_train, y_test = train_test_split(X_train_scaled, y_resampled, test_size=0.25, random_state=42)
```

```python
print("The size of training input is", X_train.shape)
print("The size of training output is", y_train.shape)
print(50 *'*')
print("The size of testing input is", X_test.shape)
print("The size of testing output is", y_test.shape)
```

```python
# Model Building
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score

from math import sqrt

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import KFold
```

```python
# 1. Linear Regression

lr = LinearRegression()
model = lr.fit(X_train,y_train)
print(f'R^2 score for train: {lr.score(X_train, y_train)}')
print(f'R^2 score for test: {lr.score(X_test, y_test)}')
```

```python
y_pred = model.predict(X_test)
```

```python
print('RMSE:',np.sqrt(mse(y_test,y_pred)))
print("MAE",mae(y_test,y_pred))
print('R-squared:',r2_score(y_test,y_pred))
```

```python
# 2. Ridge

params ={'alpha' :[0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000]}
ridge_regressor =GridSearchCV(Ridge(), params ,cv =5,scoring = 'neg_mean_absolute_error', n_jobs =-1)
ridge_regressor.fit(X_train ,y_train)
```

```python
y_train_pred =ridge_regressor.predict(X_train) ##Predict train result
y_test_pred =ridge_regressor.predict(X_test) ##Predict test result
```

```python
print("Test Results for Ridge Regressor Model:")
print(50 * '-')
print("Root mean squared error: ", sqrt(mse(y_test, y_test_pred)))
print("Mean absolute error: ",mae(y_test, y_test_pred))
print("R-squared: ", r2_score(y_test, y_test_pred))
```

```python
# 3. Dicision Tree

depth  =list(range(3,30))
param_grid =dict(max_depth =depth)
tree =GridSearchCV(DecisionTreeRegressor(),param_grid,cv =20)
tree.fit(X_train,y_train)
```

```python
y_train_pred =tree.predict(X_train) ##Predict train result
y_test_pred =tree.predict(X_test) ##Predict test result
```

```python
print("Train Results for Decision Tree Regressor Model:")
print(50 * '-')
print("Root mean squared error: ", sqrt(mse(y_train.values, y_train_pred)))
print("Mean absolute error: ", mae(y_train.values, y_train_pred))
print("R-squared: ", r2_score(y_train.values, y_train_pred))
```

```python
print("Test Results for Decision Tree Regressor Model:")
print(50 * '-')
print("Root mean squared error: ", sqrt(mse(y_test, y_test_pred)))
print("Mean absolute error: ", mae(y_test, y_test_pred))
print("R-squared: ", r2_score(y_test, y_test_pred))
```
